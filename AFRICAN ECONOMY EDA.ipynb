#### Trend Analysis

```python
def trend_analysis(econ_indicator, *countries):
    
    """
    Function to plot Trend analysis.

    Parameters:
    - econ_indicator: The economic indicator for the y-axis (e.g., 'GDP Growth')
    - countries: The country code (e.g., 'NGA' for Nigeria or more)

    Example usage:
    trend_analysis(gdp_growth, 'NGA')
    OR
    plot(gdp_growth, 'NGA', 'NER')
    """
        
    colors = ['blue','orange','green','red','purple','brown','pink','gray','olive','teal']
    
    
    fig,axs = plt.subplots(figsize=(15,10))
    
    color_ind = 0
    
    for country in countries:
        country = country.upper()
        country_dataset = dataset_cleaned[dataset_cleaned.Country == country_names[country]]
        axs.plot(country_dataset[year], country_dataset[econ_indicator], color=colors[color_ind], label=f'{country_names[country]} line plot', marker='o')
        color_ind += 1

    axs.set_xlabel('Year')
    axs.set_ylabel(econ_indicator)
    

    axs.set_title(f'Line plot: Trend Analysis on {econ_indicator} of ' + ' vs '.join(country_names[country.upper()] for country in countries))

    axs.grid(True)
    axs.legend()
  
    plt.show()
    
    fig.savefig(f"images/Trend analysis/Line plot: Trend Analysis on {econ_indicator} of {' vs '.join(country_names[country.upper()] for country in countries)}.png")
```

```python
trend_analysis(gdp_growth, 'cod')
```

### Inflation-Economic Growth Relationship

```python
def check_correlation(country):
    """
    Function that returns the correlation between Inflation and GDP Growth

    Parameters:
    - country: The country code (e.g., 'NGA' for Nigeria or more)

    Example usage:
    check_correlation('NGA')
    OR
    check_correlation('NER')
    """
    country = country.upper()
    country_dataset = dataset_cleaned[dataset_cleaned.Country == country_names[country]]
    correlation = country_dataset[inflation].corr(country_dataset[gdp_growth])
    corr = f"The correlation of {country_names[country]} between {inflation} and  {gdp_growth} is {correlation:.3}"
    return corr    
```

```python
check_correlation('NGA')
```

```python
check_correlation('STP')
```

```python
check_correlation('NER')
```

```python
corr = []
for country in country_names:
    country_dataset = dataset_cleaned[dataset_cleaned.Country == country_names[country]]
    correlation = country_dataset[inflation].corr(country_dataset[gdp_growth])
    corr.append([correlation, country_names[country]])
```

```python
sorted(corr) 
```

From the sorted correlation, we can observe that the following 

- Congo Dem. Rep has the strongest correlation between INFLATION and GDP GROWTH of -0.797
- Negative correlation shows that both ECONOMICAL INDICATOR moves inversely to each other
- As Inflation increase, GDP GROWTH decreases and vice versa

We can go as far as plot a scatter plot on it and also a regression line to further visualise it

```python
check_correlation('COD')
```

```python
def inflation_economic_plot(country):
    """
    Function that returns the correlation between Inflation and GDP Growth

    Parameters:
    - country: The country code (e.g., 'NGA' for Nigeria or more)

    Example usage:
    inflation_economic_plot('NGA')
    OR
    inflation_economic_plot('NER')
    """
    country = country.upper()
    country_dataset = dataset_cleaned[dataset_cleaned.Country == country_names[country]]
    
    fig,axs = plt.subplots(figsize=(13,13))
    
    axs.scatter(country_dataset[gdp_growth], country_dataset[inflation], color='red', label=f'Economic Relationship between {inflation} and {gdp_growth} on {country_names[country]}', marker='o')
    
    axs.set_xlabel(gdp_growth)
    axs.set_ylabel(inflation)
    axs.set_title(f'Scatter Plot: Real GDP Growth vs Inflation Trends in {country}')
    axs.grid(True)
    axs.legend()
    
    # Calculate the regression line
    regression_model = np.polyfit(country_dataset[gdp_growth], country_dataset[inflation], 1)
    print(regression_model)
    regression_line = np.poly1d(regression_model)
    print(regression_line)

    # Plot the regression line
    x_range = np.linspace(min(country_dataset[gdp_growth]), max(country_dataset[gdp_growth])+5, 100)  # Create a range of x values
    axs.plot(x_range, regression_line(x_range), color='red', linestyle='--', label='Regression Line')
    
    axs.legend()

    
    plt.show()
    
    fig.savefig(f"images/Inflation-Economic Growth Relationship/Scatter plot: inflation vs GDP Growth Analysis of {country_names[country]}.png")

    
```
