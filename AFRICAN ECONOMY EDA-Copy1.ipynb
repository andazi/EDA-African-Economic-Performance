```python
# creating a funtion that accepts 4 arguments

def plot(country = 'NGA', x_indicator = year, y_indicator = gdp_growth, chart = 'line'):
    
    """
    Function to plot economic indicators.

    Parameters:
    - country: The country code (e.g., 'NGA' for Nigeria)
    - x_indicator: The economic indicator for the x-axis (e.g., 'GDP Growth')
    - y_indicator: The economic indicator for the y-axis (e.g., 'Year')
    - chart: The type of chart (e.g., 'line', 'bar', 'scatter')

    Example usage:
    plot('NGA', year, gdp_growth, 'line')
    """
    
    country = country.upper()
    chart = chart.title()

    unique_country = dataset_cleaned[dataset_cleaned.Country == country_names[country]]
 
    x_indicator = unique_country[x_indicator]
    y_indicator = unique_country[y_indicator]
    
    
    fig,axs = plt.subplots(figsize=(15, 10)) 
    
    if chart == 'Line':
        axs.plot(x_indicator, y_indicator, color='blue', label= f'{x_indicator.name} vs {y_indicator.name} Trend', linestyle='-', marker='o')
        axs.set_xlabel(x_indicator.name)
        axs.set_ylabel(y_indicator.name)
        axs.set_title(f'Line Plot: {x_indicator.name} vs {y_indicator.name}Trends for {country_names[country]}')
        axs.grid(linestyle='--', alpha=0.6)
        axs.legend()
    elif chart == 'Bar':
        axs.bar(x_indicator, y_indicator, color='green', label= f'{x_indicator.name} vs {y_indicator.name} Trend')
        axs.set_xlabel(x_indicator.name)
        axs.set_ylabel(y_indicator.name)
        axs.set_title(f'Bar Plot: {x_indicator.name} vs {y_indicator.name} Trends for {country_names[country]}')
        axs.grid(linestyle='--', alpha=0.6)
        axs.legend()
        axs.set_xticks(x_indicator)  
        axs.set_xticklabels(x_indicator, rotation=45)  
        

    elif chart == 'Scatter':
        # Create the scatter plot
        axs.scatter(x_indicator, y_indicator, color='purple', label= f'{x_indicator.name} vs {y_indicator.name} Trend', marker='o')
        axs.set_xlabel(x_indicator.name)
        axs.set_ylabel(y_indicator.name)
        axs.set_title(f'Scatter Plot: {x_indicator.name} vs {y_indicator.name}Trends for {country_names[country]}', loc='left')
        axs.grid(True)

        axs.legend()

        # Calculate the regression line
        regression_model = np.polyfit(x_indicator, y_indicator, 1)
        regression_line = np.poly1d(regression_model)

        # Plot the regression line
        x_range = np.linspace(min(x_indicator), max(x_indicator)+5, 100)  # Create a range of x values
        axs.plot(x_range, regression_line(x_range), color='red', linestyle='--', label='Regression Line')

        axs.legend()

    else:
        print("please choose from availble plot")

    # show plot
    plt.show()
    # saving graph
    #fig.savefig(f'images/{chart} Plot: GDP Growth Trends for {country_names[country]}.png')
    
    
```

```python
# creating a funtion that accepts 3 or more arguments

def econ_comparison(country_1, country_2, x_indicator, y_indicator):
    
    """
    Function to plot economic trends.

    Parameters:
    - country_1:  country one code  (e.g., 'NGA' for Nigeria)
    - country_2:  country two code  (e.g., 'NER' for Niger)
    - x_indicator: The economic indicator for the x-axis (e.g., 'GDP Growth')
    - y_indicator: The economic indicator for the y-axis (e.g., 'Year')
  
    Example usage:
    plot('NGA', 'STP', year, inflation)
    """
    
    country_1 = country_1.upper()
    country_2 = country_2.upper()
    
    
    fig,axs = plt.subplots(figsize=(15,10))

    countries = [country_1, country_2]
    colors = ['green', 'red', 'purple']
    i = 0

    for county in countries:
        country_datset = dataset_cleaned[dataset_cleaned.Country == country_names[county]]
        axs.plot(country_datset[year], country_datset[gdp_growth], color=colors[i], label=f'{country_names[county]} line plot')
        i += 1

    axs.set_xlabel(x_indicator)
    axs.set_ylabel(y_indicator)
    axs.set_title(f'Line plot: {x_indicator} vs {y_indicator} Comparison Between {country_names[country_1]} and {country_names[country_2]}')

    axs.legend()
  

    # show plot
    plt.show()
    # saving graph
    #fig.savefig(f'images/Line plot: {x_indicator} vs {y_indicator} Comparison Between {country_names[country_1]} and {country_names[country_2]}.png')
    
    
```

```python
econ_comparison('NGA', 'STP', year, inflation)
```
